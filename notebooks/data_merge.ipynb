```python
near_mtr = pd.read_csv(data_folder % 'nearest_mtr.csv')
near_school = pd.read_csv(data_folder % 'nearest_school.csv')
near_market = pd.read_csv(data_folder % 'nearest_market.csv')

mtr_list = list(near_mtr['MTR_nearest'].values)
school_list = list(near_school['school_nearest'].values)
market_list = list(near_market['mark_nearest'].values)
estate_name_list = list(near_mtr['house_name'].values)
len_estate_name = len(estate_name_list)

# market_list
```

```python
df_new_1 = pd.DataFrame(columns = list(df_train.columns)+['dis_mtr','dis_school','dis_market'])
df_new_2 = pd.DataFrame(columns = list(df_train.columns)+['dis_mtr','dis_school','dis_market'])
df_new_3 = pd.DataFrame(columns = list(df_train.columns)+['dis_mtr','dis_school','dis_market'])
df_new_4 = pd.DataFrame(columns = list(df_train.columns)+['dis_mtr','dis_school','dis_market'])
df_new.columns

for i in tqdm(range(len_estate_name)):
    name = estate_name_list[i]
    df_tmp = df_train_estateName.loc[name].copy()
    len_tmp = len(df_tmp)
    
    if(type(df_tmp) == pd.core.series.Series):
        dict_tmp = df_tmp.to_dict()
        dict_tmp['estateName'] = name
        df_tmp = pd.DataFrame(dict_tmp,index=[0])
    else:
        df_tmp = df_tmp.reset_index()
        
    df_tmp['dis_mtr'] = pd.Series([mtr_list[i]] * len_tmp)
    df_tmp['dis_school'] = pd.Series([school_list[i]] * len_tmp)
    df_tmp['dis_market'] = pd.Series([market_list[i]] * len_tmp)

    if(i <= 1480): df_new_1 = df_new_1.append(df_tmp,sort=False)
    elif(i <= 2960): df_new_2 = df_new_2.append(df_tmp,sort=False)
    elif(i <= 4430): df_new_3 = df_new_3.append(df_tmp,sort=False)
    else: df_new_4 = df_new_4.append(df_tmp,sort=False)
        
#     if(i <= 2): df_new_1 = df_new_1.append(df_tmp,sort=False)
#     elif(i <= 4): df_new_2 = df_new_2.append(df_tmp,sort=False)
#     elif(i <= 6): df_new_3 = df_new_3.append(df_tmp,sort=False)
#     else: df_new_4 = df_new_4.append(df_tmp,sort=False)
        
# df_new_1
```

```python
df_nnnnne = df_new_1.append(df_new_2,sort=False)
df_nnnnne = df_nnnnne.append(df_new_3,sort=False)
df_nnnnne = df_nnnnne.append(df_new_4,sort=False)
df_nnnnne.isnull().any()
```
